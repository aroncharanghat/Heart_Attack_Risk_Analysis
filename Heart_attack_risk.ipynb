## Data Pre-processing  ##

```python
print('\nCount of null values in each column :\n \n',df.isnull().sum())
```

Handling missing values

Cleaning missing ages within each gender group using the median of that group.

```python
df['Age']=df.groupby('Sex')['Age'].transform(lambda x:x.fillna(x.median()))
```

Finding Height (m) from BMI and Height (cm), handling missing values with mean.

```python
def heigh(row):
    if pd.isnull(row['Height (m)']): 
        X = row['Weight (kg)'] * (row['BMI'])
        Y=math.sqrt(X)
        return Y
    else:
        return row['Height (m)']

df['Height (m)'] = df.apply(heigh, axis=1)


def heighh(row):
    if pd.isnull(row['Height (cm)']): 
        Z = row['Height (m)'] * 100
        
        return Z
    else:
        return row['Height (cm)']

df['Height (cm)'] = df.apply(heighh, axis=1)

def heiggh(row):
   
    M = row['Height (cm)'] / 100
    
    return M
    

df['Height (m)'] = df.apply(heiggh, axis=1)

mean_hei=df['Height (m)'].mean()
df['Height (m)'].fillna(mean_hei,inplace=True)
df['Height (cm)'].fillna((mean_hei * 100),inplace=True)


```

Finding Weight (kg) from BMI , handling missing values with mean.

```python
def weigg(row):
    if pd.isnull(row['Weight (kg)']): 
        X = row['BMI'] * (row['Height (m)'] ** 2)
        return X
    else:
        return row['Weight (kg)']

df['Weight (kg)'] = df.apply(weigg, axis=1)

mean_weig=df['Weight (kg)'].mean()
df['Weight (kg)'].fillna(mean_weig,inplace=True)
```

Finding BMI from Weight (kg) and Height (m).


```python
def bm(row):
    if pd.isnull(row['BMI']): 
        X = row['Weight (kg)'] / (row['Height (m)'] ** 2)
        return X
    else:
        return row['BMI']

df['BMI'] = df.apply(bm, axis=1)
```

Finding Abdominal Circumference (cm) from Height (cm) and Waist-to-Height Ratio. Handling missing values within each gender group using the median of that group.


```python
def abci(row):
    if pd.isnull(row['Abdominal Circumference (cm)']): 
        X = row['Waist-to-Height Ratio'] * (row['Height (cm)'])
        return X
    else:
        return row['Abdominal Circumference (cm)']

df['Abdominal Circumference (cm)'] = df.apply(abci, axis=1)

df['Abdominal Circumference (cm)']=df.groupby('Sex')['Abdominal Circumference (cm)'].transform(lambda x:x.fillna(x.median()))
```

Finding Waist-to-Height Ratio from Height (cm) and Abdominal Circumference (cm).

```python
def abrr(row):
    if pd.isnull(row['Waist-to-Height Ratio']): 
        X = row['Abdominal Circumference (cm)'] / (row['Height (cm)'])
        return X
    else:
        return row['Waist-to-Height Ratio']

df['Waist-to-Height Ratio'] = df.apply(abrr, axis=1)
```

Handling null values of Total Cholesterol, HDL and LDL within each gender group using the median of that group.

```python
df['Total Cholesterol (mg/dL)']=df.groupby('Sex')['Total Cholesterol (mg/dL)'].transform(lambda x:x.fillna(x.median()))

df['HDL (mg/dL)']=df.groupby('Sex')['HDL (mg/dL)'].transform(lambda y:y.fillna(y.median()))

df['Estimated LDL (mg/dL)']=df.groupby('Sex')['Estimated LDL (mg/dL)'].transform(lambda z:z.fillna(z.median()))
```

Handling null values of Fasting Blood Sugar within each Diabetes Status group using the median of that group.

```python
df['Fasting Blood Sugar (mg/dL)']=df.groupby('Diabetes Status')['Fasting Blood Sugar (mg/dL)'].transform(lambda z:z.fillna(z.median()))
```

Handling null values of CVD Risk Score within each CVD Risk Level group using the median of that group.

```python
df['CVD Risk Score']=df.groupby('CVD Risk Level')['CVD Risk Score'].transform(lambda x:x.fillna(x.median()))
```

Deleting two columns Systolic BP and Diastolic BP, these columns has many null values, we can easily derive Systolic BP and Diastolic BP from Blood Pressure column.

```python
df.drop(columns=['Systolic BP','Diastolic BP'], inplace=True)
df[['Systolic_BP', 'Diastolic_BP']] = df['Blood Pressure (mmHg)'].str.split('/', expand=True)
```

Creating derived Triglycerides column using Total Cholesterol, HDL and LDL

```python
def tri(row):
    X = row['Total Cholesterol (mg/dL)'] - row['HDL (mg/dL)']- row['Estimated LDL (mg/dL)']
    return X
    

df['Triglycerides'] = df.apply(tri, axis=1)
```
